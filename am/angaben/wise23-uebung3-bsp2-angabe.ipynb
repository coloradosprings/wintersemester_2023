## 2c. Blobs Datensatz

Verwende bei der folgenden Aufgabe nur die Features`X_blobs`, aber nicht die Labels `y_blobs`!

Der folgende Datensatz besteht aus 5 verschiedenen Cluster. Verwende zuerst `k_means` und danach ein `GaussianMixture` Modell um den Datensatz zu clustern. Vergleiche beide Methoden.

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import k_means
from sklearn.mixture import GaussianMixture
```

```python
n_clusters = 5
np.random.seed(158)
cluster_centers = (np.random.uniform(size=(n_clusters, 2)) - 0.5) * 20
cluster_std = np.random.uniform(size=(n_clusters,)) * 2
X_blobs, y_blobs = make_blobs(1000, cluster_std=cluster_std, centers=cluster_centers)

plt.scatter(X_blobs[:, 0], X_blobs[:, 1], c=y_blobs)
```
