## 2a. Moon Datensatz

Verwende bei der folgenden Aufgabe nur die Features`X_moon`, aber nicht die Labels `y_moon`!

Benutze einen DB-Scan um den Moon Datensatz in zwei Cluster zu teilen.

```python
from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN

X_moons, y_moons = make_moons(1000, noise=0.1)

```

Das folgende Bild stellt die originalen Cluster dar.

```python
plt.scatter(X_moons[:, 0], X_moons[:, 1], c=y_moons)
```

```python
# Code hier einfügen
dbs = DBSCAN(eps =0.1 , min_samples =2)
dbs.fit(X_moons)
y_pred = dbs . fit_predict (X_moons)
plt.scatter(X_moons[:, 0], X_moons[:, 1], c=y_pred)
```

## 2b. Circle Datensatz

Verwende bei der folgenden Aufgabe nur die Features`X_circ`, aber nicht die Labels `y_circ`!
- Benutze den `k_means` clustering Algorithmus um den Circle Datensatz in zwei Cluster zu unterteilen. Diese zwei Cluster müssen nicht den inneren und äußeren Kreis darstellen. 
- Können Sie eine Feature Transformation (von 2d auf 1d) finden, welche es ermöglicht den Datensatz in innerern und äußeren Kreis mittels `k_means` zu clustern?

```python
from sklearn.datasets import make_circles
from sklearn.cluster import KMeans

X_circ, y_circ = make_circles(1000, noise=0.1, factor=0.5)
plt.scatter(X_circ[:, 0], X_circ[:, 1], c=y_circ)
```

```python
k = 2
kmeans = KMeans ( n_clusters =k)
y_pred = kmeans . fit_predict (X_circ)
```

```python
plt.scatter(X_circ[:, 0], X_circ[:, 1], c=y_pred)
```

```python
def feature_map_1(X):
    return np.array([X[0],X[1],np.exp(X[1]**2+X[0]**2)])
```

```python
z  = np.array([feature_map_1(x) for x in X_circ])
```

```python
k = 2
kmeans_phi = KMeans ( n_clusters =k)
y_pred_phi = kmeans . fit_predict (z)
```

```python
plt.scatter(X_circ[:, 0], X_circ[:, 1], c=y_pred_phi)
```

## 2c. Blobs Datensatz

Verwende bei der folgenden Aufgabe nur die Features`X_blobs`, aber nicht die Labels `y_blobs`!

Der folgende Datensatz besteht aus 5 verschiedenen Cluster. Verwende zuerst `k_means` und danach ein `GaussianMixture` Modell um den Datensatz zu clustern. Vergleiche beide Methoden.

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import k_means
from sklearn.mixture import GaussianMixture
```

```python
n_clusters = 5
np.random.seed(158)
cluster_centers = (np.random.uniform(size=(n_clusters, 2)) - 0.5) * 20
cluster_std = np.random.uniform(size=(n_clusters,)) * 2
X_blobs, y_blobs = make_blobs(1000, cluster_std=cluster_std, centers=cluster_centers)

plt.scatter(X_blobs[:, 0], X_blobs[:, 1], c=y_blobs)
```

```python
k = 5
kmeans = KMeans ( n_clusters =k)
y_pred_kmeans = kmeans . fit_predict (X_blobs)
```

```python
plt.scatter(X_blobs[:, 0], X_blobs[:, 1], c=y_pred_kmeans)
```

```python
# Code hier einfügen
```

```python
from sklearn . mixture import GaussianMixture
gm = GaussianMixture ( n_components =5 , n_init =10)
gm.fit (X_blobs)
```

```python
y_pred_gaussian = gm.predict(X_blobs)
```

```python
plt.scatter(X_blobs[:, 0], X_blobs[:, 1], c=y_pred_gaussian)
```
